```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
print(tips.head())

titanic=pd.read_csv("train.csv")
# print(titanic.head())

flights=sns.load_dataset("flights")
print(flights.head())

iris=sns.load_dataset("iris")
# iris.head()
```

<h1>Multivariate Analysis</h1>
<h3>ScatterPlot (Numerical-Numerical) </h3>

```python
# bivariate analysis
sns.scatterplot(data=tips,x="total_bill",y="tip")
plt.show()

# multivariate analysis
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="sex")
plt.show()

sns.scatterplot(data=tips,x="total_bill",y="tip",hue="sex",style="smoker")
plt.show()

sns.scatterplot(data=tips,x="total_bill",y="tip",hue="sex",style="smoker",size="size")
plt.show()

# for x and y axis we have given numerical data but for hue and style we have used smoker
```

<h3>BarPlot (Numerical-Categorical) </h3>

```python
sns.barplot(data=titanic,x="Pclass",y="Age")
plt.show()

sns.barplot(data=titanic,x="Pclass",y="Fare",hue="Sex")
plt.show()
# in barplot on y axis is the mean age
```

<h3>BoxPlot (Numerical-Categorical) </h3>

```python
sns.boxplot(data=titanic,x="Pclass",y="Fare")
plt.show()

sns.boxplot(data=titanic,x="Pclass",y="Fare",hue="Sex")
plt.show()
```

<h3>DistPlot (Numerical-Categorical) </h3>

```python
sns.distplot(titanic[titanic["Survived"]==0]["Age"],hist=False)
sns.distplot(titanic[titanic["Survived"]==1]["Age"],hist=False)
```

<h3>Heatmap (Categorical-Categorical) </h3>

```python
titanic.head()
a=pd.crosstab(titanic["Pclass"],titanic["Survived"])

sns.heatmap(a)
```
