```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("../Datasets/train.csv",usecols=["Pclass","Sex","Age","Parch","Fare","Embarked","Survived"])
df.head()
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:].values,df.iloc[:,0].values, test_size=0.1, random_state=42)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

trf=ColumnTransformer(transformers=[
    ("tnf1",OneHotEncoder(),[1,5])],
    remainder="passthrough")
X_train=trf.fit_transform(X_train)
X_test=trf.fit_transform(X_test)
X_train,X_test
```

```python
from sklearn.tree import DecisionTreeRe,plot_tree
import matplotlib.pyplot as plt

dc=DecisionTreeClassifier(max_depth=3)

dc.fit(X_train,y_train)
y_pred=dc.predict(X_test)

print("Accuracy:", (y_pred == y_test).mean())
from sklearn.metrics import confusion_matrix

confusion_matrix(y_pred,y_test)

plt.figure(figsize=(12, 8))
plot_tree(dc,filled=True)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
import matplotlib.pyplot as plt

dc=DecisionTreeClassifier(max_depth=3,criterion='entropy')

dc.fit(X_train,y_train)
y_pred=dc.predict(X_test)

print("Accuracy:", (y_pred == y_test).mean())
from sklearn.metrics import confusion_matrix

confusion_matrix(y_pred,y_test)

plt.figure(figsize=(12, 8))
plot_tree(dc)
plt.show()
```
