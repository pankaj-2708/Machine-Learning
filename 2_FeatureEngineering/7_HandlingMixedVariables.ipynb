```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("titanic.csv")
df.sample(5)
```

```python
df["number"].unique()
```

```python
# for a column like number
df["number_numerical"]=pd.to_numeric(df["number"],errors='coerce',downcast='integer')
df["number_categorical"]=np.where(df["number_numerical"].isnull(),df["number"],np.nan)
```

```python
df.sample(5)
```

```python
# for a column like cabin
df["cabin_num"]=df["Cabin"].str.extract(r'(\d+)') #it extracts numerical part
df["cabin_cat"]=df["Cabin"].str[0]
```

```python
df.sample(5)
```

```python
# for column like ticket
# in this starting letters are categories and ending are numeric

df["ticket_cat"]=df["Ticket"].apply(lambda x : x.split()[0]) 
df["ticket_cat"]=np.where(df["ticket_cat"].str.isdigit(),np.nan,df["ticket_cat"])

df["ticket_num"]=df["Ticket"].apply(lambda x : x.split()[-1])
df["ticket_num"]=pd.to_numeric(df["ticket_num"],errors="coerce",downcast="integer")
```

```python
df.sample(5)
```
