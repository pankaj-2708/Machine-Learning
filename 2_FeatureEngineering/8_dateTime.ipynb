# handling date time columns

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
date=pd.read_csv("orders.csv")
time=pd.read_csv("messages.csv")
```

```python
date.info()
```

```python
date["date"]=pd.to_datetime(date["date"])
date.info()
```

```python
# extracting Year from date

date["year"]=date["date"].dt.year
date["month"]=date["date"].dt.month
date["month_name"]=date["date"].dt.month_name()
date["day"]=date["date"].dt.day
date["day_name"]=date["date"].dt.day_name()
date["day of week"]=date["date"].dt.dayofweek
date["is weekend"]=np.where(date["day_name"].isin(["Sunday","Saturday"]),1,0)
date["week of the year"]=date["date"].dt.isocalendar().week
date["quater of the year"]=date["date"].dt.quarter

import datetime

date["time elapsed"]=datetime.datetime.today()-date["date"]
date.head(5)
```

```python
time["date"]=pd.to_datetime(time["date"])
date.info()
```

```python
time["hour"]=time["date"].dt.hour
time["minute"]=time["date"].dt.minute
time["seconds"]=time["date"].dt.second
time["time"]=time["date"].dt.time
time.head()
```
