# by sckit learn

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,:4],df.iloc[:,-1],test_size=0.25)

X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import OneHotEncoder

# categories contain one d lists equal to column count in which we want to apply OridOrdinalEncoder and inside 
# those list values are give in order as their values like good is highest etc

ohe=OneHotEncoder(drop="first")

X_train_trans=ohe.fit_transform(X_train.iloc[:,2:]).toarray()
X_test_trans=ohe.fit_transform(X_test.iloc[:,2:]).toarray()
```

```python
X_train_trans=np.hstack([X_train[["brand","km_driven"]].values,X_train_trans])
X_test_trans=np.hstack([X_test[["brand","km_driven"]].values,X_test_trans])
```

```python
X_train_trans=pd.DataFrame(X_train_trans)
X_test_trans=pd.DataFrame(X_test_trans)
X_train_trans
```
