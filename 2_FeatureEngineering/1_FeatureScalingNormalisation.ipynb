```python
# it transfers data in the range of [0,1] or [-1,1]
# it is sensative to outliers

# there are four ways of doing normalisation
# 1-minmaxscaling
# 2-meanNormalisation
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Social_Network_Ads.csv")
df=df.iloc[:,:2]
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test=train_test_split(df,test_size=0.25)

X_train.shape,X_test.shape
```

#MinMaxScaling<br>

in this we convert each value to new_value=(old_value-min)/(max_value-min_value)

```python
from sklearn.preprocessing import MinMaxScaler

# it standardise by the formula given above
scaler=MinMaxScaler()

# by this command model learn the parameters 
# we only calculate parameters for x_train and then apply it to both train and test data
scaler.fit(X_train)

X_train_Scaled=scaler.transform(X_train)
X_test_Scaled=scaler.transform(X_test)

X_train_Scaled=pd.DataFrame(X_train_Scaled,columns=X_train.columns)
X_test_Scaled=pd.DataFrame(X_test_Scaled,columns=X_test.columns)
```

```python
np.round(X_test_Scaled.describe())
```

```python
np.round(X_train_Scaled.describe())
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.set_title("Before Scaling")
sns.scatterplot(x=X_test["Age"],y=X_test["EstimatedSalary"],ax=ax1)

ax2.set_title("After Scaling")
sns.scatterplot(x=X_test_Scaled["Age"],y=X_test_Scaled["EstimatedSalary"],ax=ax2)
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.set_title("Before Scaling")
sns.kdeplot(X_test["Age"],ax=ax1)
sns.kdeplot(X_test["EstimatedSalary"],ax=ax1)

ax2.set_title("After Scaling")
sns.kdeplot(X_test_Scaled["Age"],ax=ax2)
sns.kdeplot(X_test_Scaled["EstimatedSalary"],ax=ax2)
```

#Mean Normalisation<br>

in this we convert each value to new_value=(old_value-mean)/(max_value-min_value)

there is no class in sckit learn for mean normalisation

```python
def meanNorm(col):
    mean=col.mean() 
    max_val=col.max() 
    min_val=col.min() 
    tempnewcolum=[]
    for i in col:
        tempnewcolum+=[(i-mean)/(max_val-min_val)]
    return tempnewcolum
    
X_train_Scaled_Mean=pd.DataFrame({
    "Age":meanNorm(X_train["Age"]),
    "EstimatedSalary":meanNorm(X_train["EstimatedSalary"])})

X_test_Scaled_Mean=pd.DataFrame({
    "Age":meanNorm(X_test["Age"]),
    "EstimatedSalary":meanNorm(X_test["EstimatedSalary"])})
```

```python
np.round(X_train_Scaled_Mean.describe())
```

```python
np.round(X_test_Scaled_Mean.describe())
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.set_title("MinMaxScaling")
sns.scatterplot(x=X_test_Scaled["Age"],y=X_test_Scaled["EstimatedSalary"],ax=ax1)

ax2.set_title("MeanNormalisation")
sns.scatterplot(x=X_test_Scaled_Mean["Age"],y=X_test_Scaled_Mean["EstimatedSalary"],ax=ax2)
```
