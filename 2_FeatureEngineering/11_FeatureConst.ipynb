```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("../Datasets/train.csv",usecols=["Survived","Sex","Parch","SibSp"])
df.head()
```

```python
# creating a new column no of family mem by merging sibsp and parch
```

```python
df["no of family mem"]=df["SibSp"]+df["Parch"]
df
```

```python
# creating a new column family type by categorising no of family mem

def func(i):
        if i==1:
            return "alone"
        if i<3:
            return "small family"
        else:
            return "large family"

df["family type"]=df["no of family mem"].apply(func)
```

# Feature Splitting

```python
df=pd.read_csv("../Datasets/train.csv",usecols=["Name"])
df.head()
```

```python
df["first name"]=df["Name"].str.split(",",expand=True)[1].str.split(".",expand=True)[1]
df["last name"]=df["Name"].str.split(",",expand=True)[0]
```
