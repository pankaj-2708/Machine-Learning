# Function Transformer

## 1- Log Transform

it takes log of each value of data it is applied to right skewed data <br>


## 2- Reciprocal Transform

it takes reciprocal of each value of data  <br>


## 3- Square Transform

it takes square of each value of data usually applied to left skewed data  <br>

## 4- Square Root Transform

it takes square root of each value of data  <br>

NOTE - USUALLY WE APPLY EACH TRANSFORM AND CHECK WHICH GIVE BEST RESULT

transforming data also increase accuracy of some model 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats  
```

```python
df=pd.read_csv("train.csv",usecols=["Age","Fare","Survived"])
df.sample(5)
```

```python
df.isnull().sum()
```

```python
# we need to fill missing values first before transformation
df["Age"].fillna(df["Age"].mean())
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,1:3],df.iloc[:,0],test_size=0.25)
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.set_title("Before Scaling")
sns.distplot(X_train["Fare"],ax=ax1)

ax2.set_title("After Scaling")
stats.probplot(X_train["Fare"],dist="norm",plot=plt)
plt.show()
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.set_title("Before Scaling")
sns.distplot(X_train["Age"],ax=ax1)

ax2.set_title("After Scaling")
stats.probplot(X_train["Age"],dist="norm",plot=plt)
plt.show()
```

```python
# transforming

from sklearn.preprocessing import FunctionTransformer

def ApplyTransform(transformer):
    trf=FunctionTransformer(transformer)
    X_train_transformed=trf.fit_transform(X_train)
    plt.title("Before transformation Far")
    stats.probplot(X_train["Fare"],dist="norm",plot=plt)
    plt.show()
    
    plt.title("After transformation")
    stats.probplot(X_train_tranformed["Fare"],dist="norm",plot=plt)
    plt.show()

    plt.title("Before transformation Age")
    stats.probplot(X_train["Age"],dist="norm",plot=plt)
    plt.show()
    
    plt.title("After transformation")
    stats.probplot(X_train_tranformed["Age"],dist="norm",plot=plt)
    plt.show()
        
```

```python
ApplyTransform(lambda x:x**2) 
```

```python
ApplyTransform(lambda x:np.log1p) 
# log1p add 1 to all values then apply log so that log(0) problem dont arise
```
