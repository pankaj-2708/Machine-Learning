```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time

# Set up Chrome options
options = Options()
options.add_argument("--headless")  # Run in headless mode (no GUI)

# Set up the driver
driver = webdriver.Chrome(options=options)

# Open the page
driver.get("https://countries.petethompson.net/")  # Replace with your URL

# Let JavaScript load
time.sleep(3)

# Get rendered HTML
html = driver.page_source
# response=requests.get("https://countries.petethompson.net/")
# html=response.text
soup=BeautifulSoup(html,"html.parser")
# print(soup)
```

```python
countryname=soup.find_all(class_=["panel-title" ,"ng-binding"])
countryname=[i.text for i in countryname]
print(countryname)
```
