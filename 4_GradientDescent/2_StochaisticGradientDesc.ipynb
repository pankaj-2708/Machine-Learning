# Stochiastic Gradient descent

in batch gradient descent we update values of coefficent ones in an epoch

but in stochaistic gradeint descent we update values n times in an epoch (n is no of rows in data ) so it require relatively lesser no of epochs

If we go through the dataset in the same order every time, the model might learn patterns in the order, not the data.

So, we shuffle the dataset randomly in every epoch to avoid this problem — this is the core of the "randomness" in SGD.

for same no of epoch batch gradient descent is faster that SGD ( beacuse in sgd we do more updates per epoch) but batch take more epochs to reach the solution so ultimately sgd become fast

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_regression
import time
```

```python
X,y=make_regression(n_samples=100,n_features=3,n_informative=1,n_targets=1,noise=20)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lr=LinearRegression()
lr.fit(X_train,y_train)
print(lr.coef_,lr.intercept_)
y_pred=lr.predict(X_test)
print("r2 score " , r2_score(y_test,y_pred))
```

```python
class SGD:
    def __init__(self,epochs,lr):
        self.epochs=epochs
        self.lr=lr
        
    def fit(self,X,y):
        
        X=np.insert(X,0,1,axis=1)
        
        self.b=np.ones(X.shape[1])
        
        for p in range(self.epochs):
            for i in range(X.shape[0]):
                rnd=np.random.randint(0,X.shape[0])
                y_pred=X[rnd].T @ self.b  
                error = y[rnd] - y_pred 
                # print(error)
                gradient=-2 * X[rnd] * error 
                self.b=self.b - gradient * self.lr
        
        print(self.b)

    def predict(self,X):
        X=np.insert(X,0,1,axis=1)

        ans=X@self.b

        return ans
```

```python
gd=SGD(50,0.01)
st=time.time()
gd.fit(X_train,y_train)
print("Time Taken",time.time()-st)
y_pred=gd.predict(X_test)
```

```python
print("r2 score " , r2_score(y_test,y_pred))
```

```python
class GD:
    def __init__(self,epochs,lr):
        self.epochs=epochs
        self.lr=lr
        
    def fit(self,X,y):
        
        X=np.insert(X,0,1,axis=1)
        
        self.b=np.ones(X.shape[1])
        
        for p in range(self.epochs):
            y_pred=X @ self.b
            error = y - y_pred 
            gradient=-2 * X.T @ error / len(X)
            self.b=self.b - gradient * self.lr
        
        print(self.b)

    def predict(self,X):
        X=np.insert(X,0,1,axis=1)

        ans=X@self.b

        return ans
```

```python
gd=GD(50,0.01)
st=time.time()
gd.fit(X_train,y_train)
print("Time Taken",time.time()-st)
y_pred=gd.predict(X_test)
```

```python
print("r2 score " , r2_score(y_test,y_pred))
```

# Problems with SGD

1- it does not decrease steps when move toward solution so we do as below known as learning shedule

```python
def learning_rate(val):
    t0=5
    t1=50
    return t0/(val+t1)
```

```python
class SGD:
    def __init__(self,epochs,lr):
        self.epochs=epochs
        self.lr=lr
        
    def fit(self,X,y):
        
        X=np.insert(X,0,1,axis=1)
        
        self.b=np.ones(X.shape[1])
        
        for p in range(self.epochs):
            for i in range(X.shape[0]):
                rnd=np.random.randint(0,X.shape[0])
                y_pred=X[rnd].T @ self.b  
                error = y[rnd] - y_pred 
                # print(error)
                gradient=-2 * X[rnd] * error 
                self.b=self.b - gradient *  learning_rate(p*X.shape[0]+i)
        
        print(self.b)

    def predict(self,X):
        X=np.insert(X,0,1,axis=1)

        ans=X@self.b

        return ans
```

```python
gd=SGD(50,0.01)
st=time.time()
gd.fit(X_train,y_train)
print("Time Taken",time.time()-st)
y_pred=gd.predict(X_test)
print("r2 score " , r2_score(y_test,y_pred))
```
