```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("AmesHousing.csv")
df.sample(5)
```

```python
sns.scatterplot(data=df,x="Lot Area",y="SalePrice",hue="MS Zoning",style="Street",size="Land Contour")
```
