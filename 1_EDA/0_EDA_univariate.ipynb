```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("train.csv")
df.info()
# df=df[["PassengerId","Survived","Pclass","Age","SibSp","Parch","Fare"]]
```

<h1>1 Categorical Data</h1>
<h3>CountPlot</h3>

```python
sns.countplot(x=df["Survived"])
plt.show()

sns.countplot(x=df["Pclass"])
plt.show()

sns.countplot(x=df["Sex"])
plt.show()

sns.countplot(x=df["Embarked"])
plt.show()
```

<h3>PieChart</h3>

```python
df["Survived"].value_counts().plot(kind="pie",autopct='%1.2f%%')
plt.show()
# here we are using this method because for pie chart we need counts of values 
df["Pclass"].value_counts().plot(kind="pie",autopct='%1.1f%%')
plt.show()
df["Sex"].value_counts().plot(kind="pie",autopct='%1.1f%%')
plt.show()
df["Embarked"].value_counts().plot(kind="pie",autopct='%1.1f%%')
plt.show()
```

<h1>2. Numerical Data</h1>
<h3>histplot</h3>

```python
plt.hist(df["Age"],bins=8)
```

<h3>Distplot</h3>

at y axis it shows probab

```python
sns.distplot(df["Age"],bins=8)
```

<h3>Boxplot</h3>

```python
sns.boxplot(x=df["Age"])
plt.show()
```
